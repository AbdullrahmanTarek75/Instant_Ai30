## 1.1 Bernoulli

 ### Equations, And Properties:


### Explain The Function:
p is the probability of success in a Bernoulli trial. In this case, the probability of success is set to 0.3.
_______________________________________________________________________________________________________________________________________________________

bernoulli(p): creates a Bernoulli distribution object (dist) with the specified probability parameter p.
_______________________________________________________________________________________________________________________________________________________

dist.rvs(size=2): This generates an array of two random variables sampled from the Bernoulli distribution.
_______________________________________________________________________________________________________________________________________________________

The output will be something like: random_variables: [0 0] or random_variables: [1 0], where 0 and 1 represent the outcomes of Bernoulli trials (success or failure).
_______________________________________________________________________________________________________________________________________________________

x: The possible values of the random variable (0 and 1 for a Bernoulli distribution).

pmf: The probability mass function values for the corresponding values in x.

A bar plot (ax.bar) is used to visualize the PMF.

cdf: The cumulative distribution function values for the corresponding values in x.
_______________________________________________________________________________________________________________________________________________________

A scatter plot (plt.plot) with blue circles ("bo") is used to visualize the CDF.
_______________________________________________________________________________________________________________________________________________________

Vertical lines (plt.vlines) are added to connect the points in the CDF plot.



```python
##Generate the random variable
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
p=0.3
dist=bernoulli(p)
print("random_variables:",dist.rvs(size=2))
```

```python
##Plot the PMF/PDF and CDF
x=[0,1]
pmf=[dist.pmf(0),dist.pmf(1)]
fig,ax=plt.subplots()
ax.bar(x,pmf,label="PMF")
plt.xticks([0,1])
plt.title("(PMF) Probability mass function")
plt.show()
cdf=[dist.cdf(0),dist.cdf(1)]
plt.plot(x,cdf,"bo",label="CDF")
plt.vlines(x, 0, cdf, colors='b', lw=5, alpha=0.5);
plt.title("(CDF) Cumulative distribution function")
plt.show()
```

```python
##Compute The Expectation and the Variance.
print("The Expectation:",dist.mean())
print("The Variance:",dist.var())
```

## 1.2 Binomial

 ### Equations, And Properties:


### Explain The Function:
Setting Parameters: n is the number of trials in the binomial distribution, and p is the probability of success in each trial.
_______________________________________________________________________________________________________________________________________________________
Generating Random Variables: Using binom.rvs() to generate an array of 10,000 random variables following a binomial distribution with parameters n and p.
_______________________________________________________________________________________________________________________________________________________
Defining Values of r: Creating a list r_values that represents the possible values of the random variable r from 0 to n.
_______________________________________________________________________________________________________________________________________________________
Probability Mass Function (PMF) and Plotting: Calculating the Probability Mass Function (PMF) using binom.pmf() for each value in r_values and plotting the PMF using plt.bar().
_______________________________________________________________________________________________________________________________________________________
Cumulative Distribution Function (CDF) and Plotting: Calculating the Cumulative Distribution Function (CDF) using binom.cdf() for each value in r_values and plotting the CDF using plt.step().
_______________________________________________________________________________________________________________________________________________________
Printing Mean and Variance: Using binom.stats() to calculate and print the mean and variance of the binomial distribution.
_______________________________________________________________________________________________________________________________________________________


```python
##Generate the random variable
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
n = 6
p = 0.6
random_vars = binom.rvs(n, p, size=10000)
print(random_vars[:10])
```

```python
##Plot the PMF/PDF and CDF
r_values = list(range(n + 1))
pmf_values = [binom.pmf(r, n, p) for r in r_values]

# Plotting the PMF
plt.bar(r_values, pmf_values, label='PMF', alpha=0.7)
plt.title('Binomial Distribution PMF')
plt.xlabel('r')
plt.ylabel('Probability')
plt.legend()
plt.show()

# List of cdf values
cdf_values = [binom.cdf(r, n, p) for r in r_values]

# Plotting the CDF
plt.step(r_values, cdf_values, label='CDF', where='post')
plt.title('Binomial Distribution CDF')
plt.xlabel('r')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```

```python
##Compute The Expectation and the Variance.
mean, var = binom.stats(n, p)
print("Mean =", mean)
print("Variance =", var)
```

 ### Equations, And Properties:

Setting Parameters: p is set to the probability of success in a geometric distribution.
_______________________________________________________________________________________________________________________________________________________

Generating Random Variable: X is generated using np.random.geometric() with a given probability p and a sample size of 10,000.
_______________________________________________________________________________________________________________________________________________________
Computing PMF and CDF: Using the geom.pmf() and geom.cdf() functions from scipy.stats.geom to calculate the Probability Mass Function (PMF) and Cumulative Distribution Function (CDF) for the generated random variable X.
_______________________________________________________________________________________________________________________________________________________
Plotting PMF and CDF: Using matplotlib.pyplot to create a side-by-side plot of the PMF and CDF.
_______________________________________________________________________________________________________________________________________________________
•	The left subplot shows the Probability Mass Function (PMF) with unique values of X on the x-axis and their corresponding probabilities on the y-axis.
_______________________________________________________________________________________________________________________________________________________
•	The right subplot shows the Cumulative Distribution Function (CDF) with unique values of X on the x-axis and their cumulative probabilities on the y-axis.

```python
##Generate the random variable
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
p = 0.5
random_vars = geom.rvs(p, size=10000)
print(random_vars[:10])
```

```python
##Plot the PMF/PDF and CDF

# Compute the PMF and CDF of X
unique_X = np.unique(random_vars)
pmf = geom.pmf(unique_X, p)
cdf = geom.cdf(unique_X, p)

# Plot the PDF and CDF using plt.step()
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.step(unique_X, pmf, label='PDF', where='post')
plt.title('Probability Density Function (PDF)')
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()
plt.show()
plt.subplot(1, 2, 2)
plt.step(unique_X, cdf, label='CDF', where='post')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('X')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```

```python
##Compute The Expectation and the Variance.
mean = geom.mean(p)
variance = geom.var(p)
print(f'The Expectation: {mean:.2f}')
print(f'Variance: {variance:.2f}')
```

# 3.2 Predict Earthquakes (exponential distribution)

The exponential distribution is prominently used by seismologists and earth scientists to predict the approximate time when an earthquake is likely to occur in a particular locality. For this purpose, the history of the earthquakes and other natural calamities occurring in a particular locality is recorded and monitored. This data acts as the information and is fed to the exponential distribution function. The output gives an approximate time when the earthquake might occur. This helps the environmental engineers and disaster management officials draft the essential measures to avoid the loss of lives and to minimize the property destruction rate.

The probability density function (pdf) of the exponential distribution is expressed as:

f(x) = λ * exp(-λx)

Here, 'x' represents the time since the last earthquake occurred. To calculate the cumulative distribution function (cdf), the pdf is integrated over the interval [0, x]:

F(x) = ∫[0, x] λ * exp(-λt) dt

This code will generate random samples from an exponential distribution and plot the pdf and cdf of the distribution. It will also overlay a histogram of the simulated data on the plot, allowing us to compare the theoretical pdf and cdf with the empirical distribution of the data.

The results will indicate the probability of an earthquake occurring within a specific time frame, as well as the cumulative probability of multiple earthquakes occurring by a given time. This can be useful for predicting and preparing for the occurrence of natural disasters, such as earthquakes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Number of data points to generate
n = 1000
# Lambda parameter (assume an average of 2 earthquakes per year)
lambda_ = 1/2

# Generate random samples from an exponential distribution
data = expon.rvs(scale=lambda_, size=n)

# Plot the pdf 
x = np.linspace(0, 10, n )
plt.plot(x, expon.pdf(x, scale=lambda_), label='PDF')

# Overlay the histogram of the simulated data
plt.hist(data, density=True, alpha=0.5, bins=20)

plt.legend()
plt.show()
```

{1980:2.06, 1982:2, 1984:2.44,1986:1.94, 1988:1.44, 1990:1.88, 1992:1.7, 1994: 2.2, 1996:2.29, 1998:2.21, 2000:2.18, 2002:1.5, 2004:2.05, 2006:2.28, 2008:3.01, 2010:2.45, 2012:2.38, 2014:2.16, 2016:2.13, 2017:2.06, 2019:1.96}

The probability density function (pdf) of the exponential distribution is expressed as:
px(k) = P(X=k) = (e^(-λ)*( λ^k))/k!

Where

 • k = 0, 1, 2, 3... 

• e is the Euler's number (e = 2.718) 

• λ is an average rate of the expected value and λ = variance, also λ>0


 we Plot the PMF using the command The output is a Plot the PMF graphic, which shows us the different possibilities of scoring goals in a match→ We will notice that the greatest possibility is to score two goals in the match, And the lowest is score 7 This is exactly what happened in the 2021 Africa Cup, as 52 matches were scored in which 2 goals were scored, and one match in which 8 goals were scored
 


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Average Goals in each edition
average_goals = [ 1.96, 2.06, 2.13, 2.16, 2.38, 2.45, 3.01, 2.28, 2.05, 1.5, 2.18, 2.21, 2.29, 2.2, 1.7, 1.88,1.44,1.94,2.44,2,2.06]

# Calculate lambda
lambda_value = np.average(average_goals)
# Plot the PMF
x_pmf = np.arange(0, int(lambda_value * 3.5) +1)
rv_poisson = stats.poisson(lambda_value)
pmf_values = rv_poisson.pmf(x_pmf)
plt.plot(x_pmf, pmf_values, 'go')
plt.vlines(x_pmf, 0, pmf_values, 'g')
plt.title("(PMF) Probability mass function px(k)")
plt.xlabel('k')
plt.ylabel('Probability')
plt.show()
```
